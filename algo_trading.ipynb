Basic DataFrame manipulation

```python
# Data gathering
import pandas_datareader as pdr

# Data manipulation
import pandas as pd
import numpy as np
import statsmodels.api as sm
import pandas.tseries

# Data visualisation
import matplotlib.pyplot as plt

# Complements
import datetime as dt
```

```python
# Getting specific data from Stooq
msft = pd.DataFrame.from_csv("msft_us_d.csv")

# Displaying the DataFrame
msft
```

```python
# Basic informations about the dataframe

print("DataFrame infos :\n")
print(msft.describe())
print("\nLast rows :\n")
print(msft.tail()) # last rows of the dataframe (use .head() for first rows)
```

```python
# Getting label names

print("List of indexs :\n")
print(msft.index)
print("\nList of columns :\n")
print(msft.columns)
```

```python
# Subsetting a dataframe by label

close_prices = msft['Close'][-10:] # the last 10 values
print("\nSubsetting the close prices :\n")
print(close_prices) # Warning : type is Series not DataFrame
```

```python
# Subsetting with loc and iloc

start = dt.datetime(2016, 1, 1)
end = dt.datetime(2017, 1, 1)

# directly with labels and .loc
print("Subsetting with .loc :\n")
print(msft.loc[start:end].head())

# Data of the year 2014
print("\nSubsetting with .loc and year label :\n")
print(msft.loc["2014"].head())

# with indexs and .iloc
print("\nSubsetting with .iloc :\n")
print(msft.iloc[msft.index.searchsorted(start):msft.index.searchsorted(end)].head())
```

```python
# Getting a sample of the DataFrame (with 20 rows)

msft.sample(20)
```

```python
# Reshapping a DataFrame to a monthly interval

msft.resample('M').mean().head() 
# Note : 'M'-> Month ; 'Q'->Quarter ; 'A'->Annual
```

```python
# Computing the daily difference between Open and Close

msft['diff'] = msft.Open - msft.Close # Creating a new colunm
print(msft['diff'].head())
del msft['diff'] # Clearing the column
```

```python
# Plotting a DataFrame

msft['Close'].plot(grid=True, title="Close prices")
plt.ylabel("Price")
plt.show()
```

```python
# Getting daily percentage changes
daily_pct_change = msft[['Close']].pct_change()

# Filling the NAN with backward propagation
daily_pct_change.fillna(method='bfill', inplace=True)

# Plotting daily percentage change
daily_pct_change['Close'].plot(grid=True, title="Daily changes")
plt.show()
```

```python
# Plotting histogram of the returns

daily_pct_change.hist(bins=50)
plt.show()
```

```python
# Daily log returns

log_return = np.log(msft[['Close']].pct_change() + 1)
# Those returns are supposed to follow a normal law according to assumptions

log_return['Close'].plot(grid=True, title="Daily log returns")
plt.show()
```

```python
# Getting cumulative returns

cumulative_daily_return = (1 + daily_pct_change).cumprod()

cumulative_daily_return.plot(grid=True, title="Cumulative daily returns")
plt.show()
```

```python
# Quarterly returns

quarterly_return = msft.resample('Q').mean().pct_change()

quarterly_return['Close'].plot(grid=True, title="Quarterly returns")
plt.show()
```

We now want to compare different stocks between them


```python
# Getting all the datas in one DataFrame

aapl = pd.DataFrame.from_csv("aapl_us_d.csv")
ibm = pd.DataFrame.from_csv("ibm_us_d.csv")
goog = pd.DataFrame.from_csv("goog_us_d.csv")

tickers = ["AAPL", "MSFT", "IBM", "GOOG"]
datas = [aapl, msft, ibm, goog]

df = pd.concat(datas, keys=tickers, names=["Ticker", "Date"])
df
```

```python
# Computing daily percentage change
daily_close = df[['Close']].reset_index().pivot(index='Date', columns='Ticker', values='Close')
daily_pct_change = daily_close.pct_change()

# Cleaning data
daily_pct_change.dropna(axis=0, how="any", inplace=True)

# Plotting datas
daily_pct_change.hist(bins=50, sharex=True, figsize=(12, 8))
plt.show()
```

```python
# Plotting a scatter matrix of the DataFrame returns

pd.plotting.scatter_matrix(daily_pct_change, diagonal='kde', alpha=0.1, figsize=(12,12))
plt.show()
```

Computing rolling mean

```python
# Computing short and long term rolling mean
daily_close.dropna(axis=0, how="any", inplace=True)

for ticker in tickers:
    temp_str_ma_42 = ticker + "_MA_42"
    temp_str_ma_120 = ticker + "_MA_120"
    
    daily_close[temp_str_ma_42] = daily_close[[ticker]].rolling(window=42).mean()
    daily_close[temp_str_ma_120] = daily_close[[ticker]].rolling(window=120).mean()

# CLeaning data
daily_close.dropna(axis=0, how="any", inplace=True)
daily_close.head()
```

```python
# Plotting all those datas

subplot_size = int(len(tickers)/2)
fig, axarr = plt.subplots(subplot_size, subplot_size)

for index, ticker in enumerate(tickers):
    row_index = index//subplot_size
    column_index = index%subplot_size
    daily_close[[ticker, ticker+'_MA_42', ticker+'_MA_120']].plot(ax=axarr[row_index, column_index]
                                                                  , title= ticker, figsize=(12, 8))

fig.tight_layout()
plt.show()
```

```python
# Plotting the volatility of each stock

min_period = 60
volatility = daily_pct_change.rolling(min_period).std() * np.sqrt(min_period)

volatility.dropna(axis=0, how="any", inplace=True)

volatility.plot(figsize=(12, 8), title="Volatility")
plt.show()
```

```python
# OLS (Ordinary Least-Square regression)

# Calculating returns
all_close = df[['Close']]
all_returns = np.log(all_close.pct_change() + 1)

# Isolating data
aapl_returns = all_returns.iloc[all_returns.index.get_level_values('Ticker') == 'AAPL']
aapl_returns.index = aapl_returns.index.droplevel("Ticker")

msft_returns = all_returns.iloc[all_returns.index.get_level_values('Ticker') == 'MSFT']
msft_returns.index = msft_returns.index.droplevel("Ticker")

# Gathering datas
scatter_returns = pd.concat([aapl_returns, msft_returns], axis=1)
scatter_returns.columns = ['AAPL', 'MSFT']

# Cleaning data
scatter_returns.dropna(axis=0, how='any', inplace=True)

# Subsetting to the wright time period
start = dt.datetime(2016, 1, 1)
end = dt.datetime(2017, 1, 1)
scatter_returns = scatter_returns.loc[start:end]

# Building the OLS model
X = sm.add_constant(scatter_returns['AAPL'])
model = sm.OLS(scatter_returns['MSFT'], X).fit()

print(model.summary())
```

```python
# Plotting the OLS

# Scatterplot of the returns
plt.plot(scatter_returns['AAPL'], scatter_returns['MSFT'], 'b.')

# Building the regression line
x = np.linspace(min(scatter_returns['AAPL']), max(scatter_returns['AAPL']))
plt.plot(x, model.params[0] + model.params[1]*x, 'r', lw=2)

# Plot parameters
plt.grid(True)
plt.axis('tight')
plt.xlabel = 'Apple returns'
plt.ylabel = 'Microsoft returns'

# Show the graph
plt.show()
```

```python
# Checking the correlation to cross-check result

scatter_returns['MSFT'].rolling(window=60).corr(scatter_returns['AAPL']).dropna().plot()

plt.show()
```
